# Cleaning Data

## Drop Empty Column and Unnecessary Column

```python
datamain.info()
```

```python
droptable = ['webid', 'companyid', 'date_scraped', 'date_posted', 'source', 'last_seen', 'date_expired']
datamain = datamain.drop(labels=droptable, axis=1)
```

```python
datamain.info()
```

## Change Data Type

```python
# Fungsi untuk memeriksa apakah nilai adalah bilangan desimal
def is_decimal(value):
    if isinstance(value, float):
        return value % 1 != 0 and value % 0.1 != 0 and value != 0
    return False

# Memeriksa kolom 'A' untuk keberadaan bilangan desimal
decimal_values_min = datamain['salary_min'].apply(lambda x: is_decimal(x))
decimal_values_max = datamain['salary_max'].apply(lambda x: is_decimal(x))

datamain['Decimal with Nonzero min'] = decimal_values_min
datamain['Decimal with Nonzero max'] = decimal_values_min

# Menampilkan DataFrame
datamain.head(20)
```

```python
int_columns = ["year_experience_min", "year_experience_max", "salary_min", "salary_max", "remote"]

for column in int_columns:
    datamain[column] = pd.to_numeric(datamain[column], errors='coerce').fillna(0).astype(int)
```

```python
datamain.info()
```

```python
datamain.nunique()
```

```python
column = ["Decimal with Nonzero min", "Decimal with Nonzero max"]
datamain.drop(labels=column, axis=1, inplace=True)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter

# Menggabungkan semua teks menjadi satu teks
all_text = ' '.join(datajob['responsibility'])

# Memisahkan teks menjadi token (kata-kata atau kalimat)
tokens = word_tokenize(all_text)

# Menghitung frekuensi masing-masing token
token_counter = Counter(tokens)

# Dapatkan 50 kata paling banyak digunakan
top_100_words = token_counter.most_common(100)

# Filter baris-baris yang mengandung salah satu dari 100 kata paling umum
filtered_data = datajob[datajob['responsibility'].str.contains('|'.join([word for word, count in top_100_words]))]

# Tampilkan hasil
print("100 kata paling banyak digunakan:", top_100_words)
```

```python
# Analyze the responsibilities
responsibilities = datajob['responsibility']

# Identify potential gig jobs based on keywords
gig_jobs_keywords = ['social media management', 'content creation', 'digital marketing', 'email marketing', 'SEO optimization', 'advertising', 
                     'Assistant Manager', 'Campaign management', 'Market research', 'Brand management', 'Public relations', 'Sales and marketing',
                     'Lead generation', 'Marketing strategy', 'Market segmentation', 'Competitor analysis']

# Membuat DataFrame dari list
df_keywords = pd.DataFrame({'Gig Jobs Keywords': gig_jobs_keywords})

# Buat tabel
# Inisialisasi DataFrame untuk hasil
result_df = pd.DataFrame(columns=['Gig Jobs Keywords', 'responsibility'])

# Loop melalui setiap kata kunci
for keyword in df_keywords['Gig Jobs Keywords']:
    # Gunakan str.contains untuk mencari kemunculan kata kunci dalam setiap kalimat
    mask = datajob['responsibility'].str.contains(keyword, case=False)
    # Filter kalimat yang mengandung kata kunci
    sentences_with_keyword = datajob[mask]
    # Tambahkan hasil pencarian ke dalam DataFrame hasil
    sentences_with_keyword['Gig Jobs Keywords'] = keyword
    result_df = pd.concat([result_df, sentences_with_keyword[['Gig Jobs Keywords', 'responsibility']]], ignore_index=True)

# Kelompokkan kalimat berdasarkan kata kunci
grouped_df = result_df.groupby('Gig Jobs Keywords')['responsibility'].apply(list).reset_index()

# Hitung jumlah responsibilitas yang mengandung setiap kata kunci
grouped_df['Total'] = grouped_df['responsibility'].apply(lambda x: len(x))

# Tampilkan hasil
grouped_df.sort_values(by='Total', ascending=False)
```
